```python
sp = 0.22 #Aksjeutbytter og gevinster over skjermingsbeløp skattlegges som alminnelig inntekt, med 22 prosent skatteåret 2021 og 2022. 
k = 100000 # skjermingsgrunnlag
sb = 3800 # Ubenyttet skjermingsbeløp
g1 = 1.44 #I 2021 skal utbytte og gevinster på aksjer og aksjefond ganges opp med en faktor på 1,44, slik at skattesatsen i praksis blir 31,68 prosent. 
g2 = 1.60 # I 2022 økes denne oppjusteringsfaktoren til 1,60, alik at skattesatsen i praksis blir 35,20 prosent.
n = 120000 #Aksjens salgspris
ag = n-k #Aksjens gevinst
aog = (n-k-sb)*g1 #Aksjens oppjusterte gevinst
skatt = aog*sp
resultat = ag-skatt

print(ag, skatt, resultat)
```
