```python
import sys
import os
import pandas as pd
from IPython.display import display

CURR_DIR = os.getcwd()
print(os.path.basename(CURR_DIR))
print(CURR_DIR)
```

```python
oslo_bors_df=pd.read_csv(CURR_DIR+"/symboler_oslo_bors.csv").drop(['ISIN', 'Market'], axis=1)
oslo_bors_df.to_csv("name_symbol_all_oslo_bors.txt", index=False)
```

```python
portfolio_df=pd.DataFrame([],columns=["Name", "Symbol"])
candidates_df=pd.DataFrame([],columns=["Name", "Symbol"])
portfolio_df=pd.read_csv(CURR_DIR+"/portfolio.txt")
candidates_df=pd.read_csv(CURR_DIR+"/candidates.txt")
portfolio_df.sort_values(by=['Name']).reset_index(drop=True)
candidates_df.sort_values(by=['Name']).reset_index(drop=True)
print("Portfolio")
display(portfolio_df)
print("Candidates")
display(candidates_df)
```
